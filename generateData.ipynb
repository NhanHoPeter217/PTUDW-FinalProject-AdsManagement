```python
import random as rd
import json
```

```python
infos = ['Cổ động chính trị',
         'Thực phẩm',
         'Dịch vụ',
         'Giải trí',
         'Thể thao',
         'Nghỉ dưỡng',
         'Vật dụng',
         'Y tế'
        ]

types = ['Chuyển hình Rolling',
         'Biển quảng cáo động Trivision',
         'Biển quảng cáo điện tử',
         'Biển quảng cáo đèn LED'
         'Pano ốp tường',
         'Billboard - Biển lớn một cột',
         'Lightbox - Biển hộp đèn'
        ]
```

```python
def randType():
    return rd.choice(types)

def randInfo():
    return rd.choice(infos)

def randSize(place = ''):
    w_min = 1.5
    w_max = 10

    h_min = 1.5
    h_max = 10
    
    if (place == types[5]):
        w_min = 10
        w_max = 15

        h_min = 15
        h_max = 20
    
    return {"w": rd.random() * (w_max - w_min) + w_min,
            "h": rd.random() * (h_max - h_min) + h_min
            }

def randStatus():
    return rd.choices([0,1], [0.2, 0.8])[0]
```

```python
class coords:
    def __init__(self, lat = 1.0, lng = 1.0):
        self.lat =  lat
        self.lng = lng

class AdvertisementBoard:
    def __init__(self, title = '', type = '', info = '', coords = any, address = '', size = {"w": 1.0, "h": 1.0}, n = 1, status = 0) -> None:
        self.title = title
        self.type = type
        self.info = info
        self.coords = coords
        self.address1 = address
        self.size = size
        self.status = status
        self.n = n
```

```python
def generateJSON():
    x = []
    for location in inits:
        type = randType()
        x.append(AdvertisementBoard(
            title = location["title"],
            type = type,
            info = randInfo(),
            coords = coords(float(location['lat']), float(location['lng'])),
            status= randStatus(),
            size = randSize(type),
            n= rd.randint(1,5)
        ))
    return x
```

```python
class MyEncoder(json.JSONEncoder):
        def default(self, o):
            return o.__dict__

x = generateJSON()

with open('public\js\locations.js', 'w') as fw:
    fw.write('const locations = ')
    fw.write(MyEncoder().encode(x))
    fw.write(';\nexport default locations;')
```
